## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files

mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Display the data table for preview
merge_df.head()
```

```python
# Checking the number of mice.
total_mice = merge_df["Mouse ID"].nunique()
print(total_mice)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dup = merge_df[merge_df.duplicated(subset=["Mouse ID","Timepoint"], keep=False)]
dupname = dup["Mouse ID"].value_counts()
dupname
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
merge_df[merge_df.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merge_df.loc[merge_df["Mouse ID"] != "g989"]
clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice_df = clean_df["Mouse ID"].nunique()
clean_mice_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

# Variables for statistical information
mean = clean_df[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Drug Regimen").mean()
median = clean_df[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Drug Regimen").median()
variance = clean_df[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Drug Regimen").var()
standard = clean_df[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Drug Regimen").std()
SEM = clean_df[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Drug Regimen").sem()

groupby_df = clean_df.groupby(["Drug Regimen"]).count()

summary_df = pd.DataFrame(groupby_df)
summary_df["mean"] = mean
summary_df["median"] = median
summary_df["var"] = variance
summary_df["std"] = standard
summary_df["sem"] = SEM
summary_stat_df = summary_df[["mean", "median", "var", "std", "sem"]]
summary_stat_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line
clean_agg_df = clean_df[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Drug Regimen").agg({"Tumor Volume (mm3)": ['mean', 'median', 'var', 'std', 'sem']})
clean_agg_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
filt_clean_df = clean_df[(clean_df["Drug Regimen"] == "Capomulin")|
                         (clean_df["Drug Regimen"] == "Ramicane") |
                         (clean_df["Drug Regimen"] == "Infubinol") |
                         (clean_df["Drug Regimen"] == "Ceftamin") ]
maxtime_df = filt_clean_df[["Mouse ID", "Drug Regimen", "Timepoint"]].groupby(["Drug Regimen", "Mouse ID"]).max()
maxtime_df

newmerge_df = pd.merge(clean_df,maxtime_df,on="Mouse ID")
merges_df = newmerge_df.rename(columns={"Mouse ID": "Mouse ID", "Drug Regimen": "Drug Regimen","Timepoint_x": "Original Timepoint",
                                          "Timepoint_y":"Final Timepoint","Weight(g)":"Weight(g)", "Tumor Volume (mm3)": "Final Tumor Volume (mm3)"})
merges_df
reseti_filtclean_df = merges_df[(merges_df["Original Timepoint"] == merges_df["Final Timepoint"])]
reseti_filtclean_df
```

```python
capo_df = reseti_filtclean_df[(reseti_filtclean_df["Drug Regimen"] == "Capomulin")]
capo_df
ram_df = reseti_filtclean_df[(reseti_filtclean_df["Drug Regimen"] == "Ramicane")]
ram_df
inf_df = reseti_filtclean_df[(reseti_filtclean_df["Drug Regimen"] == "Infubinol")]
inf_df
cef_df = reseti_filtclean_df[(reseti_filtclean_df["Drug Regimen"] == "Ceftamin")]
cef_df
```

```python
# # Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
```

```python
# Capomulin Quartiles
capo_quartiles = capo_df["Final Tumor Volume (mm3)"].quantile([.25,.5,.75])
capo_upperquartile = capo_quartiles[0.75]
capo_lowerquartile = capo_quartiles[0.25]  
print("Capomulin Information:")
print("- - - - - - - - - - - - - - -")
print(f"Quartiles are as follows: {capo_quartiles[.25]},{capo_quartiles[.5]},{capo_quartiles[.75]}")
# Capomulin IQR
capo_iqr = capo_upperquartile-capo_lowerquartile
print("- - - - - - - - - - - - - - -")
print(f"InterQuartile Range = {capo_iqr}")
print("- - - - - - - - - - - - - - -")

# Capomulin Outlier Boundaries
capo_upperboundary = capo_upperquartile + (1.5*capo_iqr)
capo_lowerboundary = capo_lowerquartile - (1.5*capo_iqr)
print(f"Top boundary = {capo_upperboundary}")
print(f"Low boundary = {capo_lowerboundary}")
print("- - - - - - - - - - - - - - -")

# Capomulin Potential Outliers
capo_out = capo_df.loc[(capo_df["Final Tumor Volume (mm3)"] < capo_lowerboundary) | (capo_df["Final Tumor Volume (mm3)"] > capo_upperboundary)]
print(f"Ceftamin outliers are: \n{capo_out}")
print("- - - - - - - - - - - - - - -")
print(f"Anything below {capo_lowerboundary} or above {capo_upperboundary} is considered a potential outlier")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
cap_val = reseti_filtclean_df.loc[reseti_filtclean_df["Drug Regimen"] == "Capomulin", ["Final Tumor Volume (mm3)"]]
cap_val = cap_val["Final Tumor Volume (mm3)"].tolist()
ram_val = reseti_filtclean_df.loc[reseti_filtclean_df["Drug Regimen"] == "Ramicane", ["Mouse ID", "Final Tumor Volume (mm3)"]]
ram_val = ram_val["Final Tumor Volume (mm3)"].tolist()
inf_val = reseti_filtclean_df.loc[reseti_filtclean_df["Drug Regimen"] == "Infubinol", ["Mouse ID", "Final Tumor Volume (mm3)"]]
inf_val = inf_val["Final Tumor Volume (mm3)"].tolist()
cef_val = reseti_filtclean_df.loc[reseti_filtclean_df["Drug Regimen"] == "Ceftamin", ["Mouse ID", "Final Tumor Volume (mm3)"]]
cef_val = cef_val["Final Tumor Volume (mm3)"].tolist()

drugs = [cap_val, ram_val, inf_val, cef_val]

red_diamond = dict(markerfacecolor='r', marker='D')

fig1, ax1 = plt.subplots()
ax1.set_title('Tumor Volumes by Drug Regimen')
ax1.set_ylabel('Final Tumor Volume')
ax1.set(ylim=(20, 75))
ax1.boxplot(drugs, whis = 1.5, showfliers = True, flierprops=red_diamond, labels = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"])
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
correlation = st.pearsonr(capo_df["Final Tumor Volume (mm3)"],capo_df["Weight (g)"])
print(f"The correlation coefficient between TUMOR VOLUME and MOUSE WEIGHT is {round(correlation[0],2)}")               
# for mouse weight and average tumor volume for the Capomulin regimen
```

```python
x = capo_df["Final Tumor Volume (mm3)"]
y = capo_df["Weight (g)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
regress = x * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
ax1 = plt.subplots(figsize=(15, 10))
plt.scatter(x,y,s=400)
plt.plot(x,regress,"r-",linewidth=6)
plt.annotate(line_eq,(25,24),fontsize=25,color="red")
plt.xlabel('Final Tumor Volume',fontsize=25)
plt.ylabel('Mouse Weight',fontsize=25)
plt.title("Linear Regression Model: Weight vs. Tumor Volume", fontsize=25)
plt.show()
```
